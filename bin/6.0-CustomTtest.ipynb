# Search contigs and run T test

Searchlist is composed of a list of search commands "HAS", "rank", "name"

HAS is a bool that when set to true makes the search.

rank is what level of taxonomy that you want to search under.

name is the name of that rank that you want to return.

--------------------------------------------------------------------------

So if you wanted to search for the species "Abacarus hystrix" the search command would be...

searchlist = [(True,"species","Abacarus hystrix")]

If you wanted to remove all the superkingdoms under "NA" you would do this

searchlist = [(False,"superkingdom","NA")]

Search commands also can be trained together like this...

searchlist = [(False,"superkingdom","NA"), (True,"species","Abacarus hystrix")]

```python
# This will run all contigs 
searchlist = None
binRank = None
control = conditions[0][0]
treatment = conditions[0][1]
title = "OUT_All_Contigs"
title = f"{control}_{treatment}_OUT_ALL_CONTIGS"

T = FilterTtest(NF_out=NF_out, 
             NF_path=NF_path,
             col_data = col_data, 
             title=title, 
             searchlist=searchlist,
             control = control, 
             treatment = treatment, 
             binRank = binRank,
             min_length = 100,
             p_adj = 0.05
            )

T.runSearch()
```


# Run the same with a permutation of the conditions

# Bin by rank

This will bin the counts by rank and then run the T test

```python


for permutation in conditions:
    control = permutation[0]
    treatment = permutation[1]

    for superkingdom in superkingdoms:
        print(f"**** working on {superkingdom}")
        searchlist = [(True,"superkingdom",superkingdom)]
        for binRank in ranks:

            makeFolders(f"OUT_{superkingdom}")
            title = f"{control}_{treatment}_{superkingdom}_Bin_{binRank}"

            T = FilterTtest(NF_out=NF_out, 
                         NF_path=NF_path,
                         col_data = col_data, 
                         title=title, 
                         searchlist=searchlist,
                         control = control, 
                         treatment = treatment, 
                         binRank = binRank,
                         min_length = 100,
                         p_adj = 0.05
                        )

            T.runSearch()
```
